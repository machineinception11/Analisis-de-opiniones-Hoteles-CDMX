```python
from time import sleep
import random
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
```

```python
# Función para hacer scrolling
scrollingScript = """
    document.getElementsByClassName('uitk-dialog-content default-padding')[0].scroll(0,2000000)
"""
# Diccionario donde guardaremos los datos
opiniones = {"usuario": [], "fecha": [], "calificacion": [], "calificacion": []}

# Cambiamos los headers para tener un poco de seguridad
options = Options()
options.add_argument("user_agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36")

# Indicamos donde está el driver de chrome
driver = webdriver.Chrome(r'C:\Users\HP-001\Aprendiendo\Web scraping\chromedriver.exe', chrome_options = options)

# ¿De donde extraeremos los datos?
driver.get("https://www.expedia.mx/Mexico-City-Hoteles-Gran-Hotel-Ciudad-De-Mexico.h21358.Informacion-Hotel?chkin=2021-05-22&chkout=2021-05-23&destType=CURRENT_LOCATION&destination=Ciudad%20de%20M%C3%A9xico%20%28y%20alrededores%29%2C%20Ciudad%20de%20M%C3%A9xico%2C%20M%C3%A9xico&neighborhoodId=3000662955&pwaDialog=sui-see-all-reviews-dialog&pwa_ts=1620520797949&referrerUrl=aHR0cHM6Ly93d3cuZXhwZWRpYS5teC9Ib3RlbC1TZWFyY2g%3D&regionId=178285&rfrr=HSR&rm1=a2&selected=21358&selectedRatePlan=206180519&selectedRoomType=507872&semdtl=&sort=RECOMMENDED&top_cur=MXN&top_dp=2312&useRewards=false&userIntent=&x_pwa=1")

# Esperamos a que la página cargue
# WebDriverWait(driver, 10).until(
#     EC.visibility_of_all_elements_located((By.XPATH, '//button[contains(@data-photo-index, "3")]'))
# )
sleep(random.uniform(1,2))

boton1 = driver.find_element(By.XPATH, "//div[@itemprop = 'aggregateRating']/div/div/button")
boton1.click()

sleep(random.uniform(1,2))
SCROLLS = 0

# Realizar primero los scrolls para cargar todas las optiniones que querramos
while (True):
    try:
        # Ejecutamos a partir de una funciónde JavaScript el scrolling
        driver.execute_script(scrollingScript)
        # Esperamos a que se haya cargado la página
        sleep(random.uniform(0,1))
        # Le sumamos 1 al contador de scrolling
        SCROLLS += 1
        boton2 = driver.find_element(By.XPATH, "//div[@data-stid='property-reviews-list']/section/div[2]/button")
        boton2.click()
    except:
        break
      
    
# Empezamos a ingresar a cada opinion y extraer los datos
contenedor = driver.find_elements_by_xpath('//div[@class = "uitk-card-aloha-content-section uitk-card-aloha-content-section-border-bottom uitk-card-aloha-content-section-padded"]/article[@itemprop="review"]')

data = {"calificaciones" : [], "fechas": [], "reseñas": [], "autor": []}


for element in contenedor:
    calificacion = element.find_element(By.XPATH, './/h4/span[@itemprop="ratingValue"]').get_attribute('aria-label')
    fecha = element.find_element_by_xpath('.//div[@class = "uitk-type-300 uitk-text-default-theme"]/span').text
    try:
        reseña = element.find_element_by_xpath('.//span[@itemprop="description"]').text
    except:
        reseña = "Sin reseña"
    autor = element.find_element_by_xpath('.//h4[@itemprop="author"]').text
    
    data["calificaciones"].append(calificacion)
    data["fechas"].append(fecha)
    data["reseñas"].append(reseña)
    data["autor"].append(autor)
```

```python
data = pd.DataFrame(data)
```
