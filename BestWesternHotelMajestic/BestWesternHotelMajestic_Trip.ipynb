# Extracción de Opiniones Hotel 2.	Best Western Hotel Majestic ($960)

```python
# Importamos librerias
from time import sleep
import pandas as pd
import random
from scrapy.item import Item
from scrapy.item import Field
from scrapy.spiders import CrawlSpider, Rule
from scrapy.loader import ItemLoader
from scrapy.linkextractors import LinkExtractor
from scrapy.selector import Selector
from scrapy.crawler import CrawlerProcess
```

```python
class Opiniones(Item):
    nombre = Field()
    opinion = Field()
    calificacion = Field()
    fecha = Field()
    
class Hotel(CrawlSpider):
    # Definimos el nombre del spider
    name = "BestWesternHotelMajestic"
    
    # Definimos la seguridad para no sér detectados
    custom_settings = ({
        "USER_AGENT": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"
    })
    
    # Tiempo de descarga
    download_delay = 1
    
    # ¿De dónde extraerémos los datos?
    start_urls = ["https://www.tripadvisor.com.mx/Hotel_Review-g150800-d152638-Reviews-or355-Best_Western_Hotel_Majestic-Mexico_City_Central_Mexico_and_Gulf_Coast.html"]
    
    # ¿Qué páginas están permitidas?
    allowed_domain = ["tripadvisor.com.mx"]
    
    # Definimos las reglas
    # Scraping horizontal
    rules = (
        Rule(
            LinkExtractor(
                allow = r'-or\d+'
            ), follow = True, callback = 'parse_opiniones'
        ),
    )
    
    # Definimos la función para el parseador
    def parse_opiniones(self, response):
        sel = Selector(response)
        # opiniones = sel.xpath('//div[@data-test-target="reviews-tab"]/div[@data-test-target = HR_CC_CARD]')
        opiniones = sel.xpath('//div[@id = "taplc_resp_hr_ad_wrapper_lower_block_0"]/div/div/div/div/div/div[3]/div[@data-test-target="HR_CC_CARD"]')
        
        for opinion in opiniones:
            item = ItemLoader(Opiniones(), opinion)
            item.add_xpath('nombre', './/a[@class = "ui_header_link _1r_My98y"]/text()')
            item.add_xpath('opinion', './/q[@class = "IRsGHoPm"]/span/text()') 
            item.add_xpath('calificacion', './/div[@data-test-target = "review-rating"]/span[contains(@class, "ui_bubble_rating")]/@class')
            item.add_xpath('fecha', './/span[@class = "_34Xs-BQm"]/text()') 
            
            yield item.load_item()

```

```python
process = CrawlerProcess({
    "FEED_FORMAT": "csv",
    "FEED_URI": "BestWesternHotelMajesticc_Trip.csv"
})

process.crawl(Hotel)
process.start()
```
